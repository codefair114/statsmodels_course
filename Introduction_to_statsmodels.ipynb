```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = sm.datasets.get_rdataset("Guerry", "HistData").data
```

```python
data
```

```python
# Fit OLS regression using ln with R-style formulas
res_r = smf.ols("Lottery ~ Literacy + np.log(Pop1831)", data=data).fit()
```

```python
res_r.summary()
```

```python
# Fit with numpy arrays
nobs = 100
X = np.random.random((nobs, 2))
X = sm.add_constant(X)
```

```python
beta = [1,.1,.5] #coef
e = np.random.random(nobs) #error
y = np.dot(X, beta) + e
```

```python
res_np = sm.OLS(y, X).fit()
```

```python
res_np.summary()
```
